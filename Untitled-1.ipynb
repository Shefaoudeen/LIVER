```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, Input, Add
import cv2
import tensorflow as tf
import random

# Clear GPU memory and set memory growth to avoid errors
tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)

# Set TensorFlow to use CPU only to avoid GPU memory errors
tf.config.set_visible_devices([], 'GPU')

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

# Load the grayscale image
img_path = '/home/icmr/Pictures/CT.jpeg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Enhance the image using histogram equalization
img_enhanced = cv2.equalizeHist(img)

# Check the enhanced image dimensions
height, width = img_enhanced.shape
print(f"Enhanced Grayscale Image - Width: {width}, Height: {height}")

# Resize the image to smaller size (optional step to reduce memory consumption)
resized_img = cv2.resize(img_enhanced, (128, 128))  # Resize to smaller dimensions
resized_height, resized_width = resized_img.shape

# Normalize the resized image (scale pixel values to range [0, 1])
resized_img_normalized = resized_img / 255.0

# Add an extra dimension to represent the batch size (needed for Keras)
# Since the image is grayscale, we only have one channel
img_batch = np.expand_dims(np.expand_dims(resized_img_normalized, axis=-1), axis=0)

# Define the input layer
input_layer = Input(shape=(resized_height, resized_width, 1))

# Apply the first Conv2D layer with 32 filters and a residual connection
conv_layer1 = Conv2D(32, (3, 3), activation='relu', padding='same')(input_layer)
residual_output1 = Add()([input_layer, conv_layer1])

# Apply the second Conv2D layer with 64 filters and a residual connection
conv_layer2 = Conv2D(64, (3, 3), activation='relu', padding='same')(residual_output1)
shortcut2 = Conv2D(64, (1, 1), activation=None, padding='same')(residual_output1)  # Adjust channels
residual_output2 = Add()([shortcut2, conv_layer2])

# Apply the third Conv2D layer with 128 filters and a residual connection
conv_layer3 = Conv2D(128, (3, 3), activation='relu', padding='same')(residual_output2)
shortcut3 = Conv2D(128, (1, 1), activation=None, padding='same')(residual_output2)  # Adjust channels
residual_output3 = Add()([shortcut3, conv_layer3])

# Apply the fourth Conv2D layer with 256 filters and a residual connection
conv_layer4 = Conv2D(256, (3, 3), activation='relu', padding='same')(residual_output3)
shortcut4 = Conv2D(256, (1, 1), activation=None, padding='same')(residual_output3)  # Adjust channels
residual_output4 = Add()([shortcut4, conv_layer4])

# Create the model
model = Model(inputs=input_layer, outputs=[residual_output1, residual_output2, residual_output3, residual_output4])

# Get the filtered outputs from the model (including the residual connections)
filtered_imgs = model.predict(img_batch)

# Remove the batch dimension from each output
filtered_imgs = [filtered_img[0] for filtered_img in filtered_imgs]

# Display the original, enhanced, and filter layer outputs
plt.figure(figsize=(20, 10))

# Original grayscale image
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')  # Display as grayscale
plt.title('Original Grayscale Image')
plt.axis('off')

# Enhanced grayscale image
plt.subplot(2, 3, 2)
plt.imshow(img_enhanced, cmap='gray')  # Display enhanced image
plt.title('Enhanced Grayscale Image')
plt.axis('off')

# Display each filter layer output
filter_titles = ['32 Filters', '64 Filters', '128 Filters', '256 Filters']
for i, filtered_img in enumerate(filtered_imgs):
    summed_filter_img = np.sum(filtered_img, axis=-1)  # Sum all feature maps
    plt.subplot(2, 3, i + 3)
    plt.imshow(summed_filter_img, cmap='gray')
    plt.title(f'Filter Layer ({filter_titles[i]})')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
