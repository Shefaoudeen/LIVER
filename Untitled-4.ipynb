```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
import cv2
import tensorflow as tf

# Force TensorFlow to use CPU instead of GPU
tf.config.set_visible_devices([], 'GPU')

# Load the grayscale image
img_path = '/home/icmr/Pictures/CT.jpeg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read image in grayscale

# Get the original image dimensions
height, width = img.shape
print(f"Original Grayscale Image - Width: {width}, Height: {height}")

# Normalize the image (scale pixel values to range [0, 1])
img_normalized = img / 255.0

# Add an extra dimension to represent the batch size (needed for Keras)
# Since grayscale image has one channel, the shape is (height, width, 1)
img_batch = np.expand_dims(np.expand_dims(img_normalized, axis=-1), axis=0)

# Create a simple model with 32 filters using Conv2D
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(height, width, 1))
])

# Get the filtered output from the model (the convolution layer)
filtered_img = model.predict(img_batch)

# Remove the batch dimension to get the final image shape
filtered_img = filtered_img[0]

# Create a grid of subplots to display the feature maps
num_filters = filtered_img.shape[2]
cols = 8  # Number of columns in the grid (can be adjusted)
rows = num_filters // cols  # Number of rows based on the number of filters
if num_filters % cols != 0:
    rows += 1  # Add an extra row if the number of filters is not divisible by columns

# Display the original image and feature maps in a grid
plt.figure(figsize=(15, 15))

# Display the original image
plt.subplot(rows + 1, cols, 1)
plt.imshow(img, cmap='gray')  # Display as grayscale
plt.title('Original Grayscale Image')
plt.axis('off')

# Display each feature map in the grid
for i in range(num_filters):
    plt.subplot(rows + 1, cols, i + 2)  # Start at index 2 to leave space for the original image
    plt.imshow(filtered_img[:, :, i], cmap='viridis')  # Display feature map using a colormap
    plt.title(f'Feature Map {i + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
